```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_wine_congr = pd.read_csv("../data/results_wine_congVoting.csv")
```

```python
df_wine_congr
```

```python
df_reviews = pd.read_csv("../data/results_reviews.csv")
```

```python
df_reviews
```

```python
df_bank = pd.read_csv("../data/results_bank_marketing.csv")
```

```python
df_bank
```

```python
combined_results = pd.concat([df_wine_congr, df_reviews, df_bank], ignore_index=True)
```

```python
combined_results
```

```python
combined_results['feature_selection'].fillna("none", inplace=True)
combined_results['scaling'].fillna("none", inplace=True)
```

```python
combined_results.to_csv('../data/combined_results.csv', index=False)
```

```python
combined_results
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='f1', hue='classifier', data=combined_results)
plt.title('Comparison of f1 across datasets and classifiers')
plt.ylabel('F1-score')
plt.xlabel('Datasets')
plt.savefig('f1_datasets_classifiers.png')
plt.show()
```

```python
#plot f1-score across datasets for each classifier, cv vs holdout
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='f1', hue='splitting', data=combined_results)
plt.title('Mean F1-score by splitting strategy per dataset')
plt.ylabel('F1-score')
plt.xlabel('Datasets')
plt.savefig('f1_datasets_splitting.png')
plt.show()
```

```python
# data.groupby(['dataset', 'Spliting'])['Mean F1-score'].mean().unstack().plot(kind='bar', figsize=(10,5))
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='f1', hue='feature_selection', data=combined_results)
plt.title('Mean F1-score by feature selection method per dataset')
plt.ylabel('F1-score')
plt.xlabel('Datasets')
plt.savefig('f1_datasets_feature_selection.png')
plt.show()
```

```python
filtered_results = combined_results[combined_results['dataset'] != 'bank_marketing']

plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='f1', hue='scaling', data=filtered_results)
plt.title('Mean F1-score by scaling method per dataset')
plt.ylabel('F1-score')
plt.xlabel('Datasets')
plt.savefig('f1_datasets_scaling.png')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='training_time', hue='classifier', data=combined_results, log=True)
plt.title('Comparison of the training time across datasets and classifiers')
plt.ylabel('Training Time')
plt.xlabel('Datasets')
plt.savefig('trainingtime_datasets_classifiers.png')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='preprocess_time', hue='classifier', data=combined_results, log=True)
plt.title('Comparison of the preprocessing time across datasets and classifiers')
plt.ylabel('Preprocessing Time')
plt.xlabel('Datasets')
plt.savefig('preprocessingtime_datasets_classifiers.png')
plt.show()
```
