```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_wine_congr = pd.read_csv("../data/results_wine_congVoting.csv")
```

```python
df_wine_congr
```

```python
df_reviews = pd.read_csv("../data/results_reviews.csv")
```

```python
df_reviews
```

```python
df_bank = pd.read_csv("../data/results_bank_marketing.csv")
```

```python
df_bank
```

```python
combined_results = pd.concat([df_wine_congr, df_reviews, df_bank], ignore_index=True)
```

```python
combined_results
```

```python
combined_results['feature_selection'].fillna("none", inplace=True)
combined_results['scaling'].fillna("none", inplace=True)
```

```python
combined_results.to_csv('../data/combined_results.csv', index=False)
```

```python
combined_results
```

```python
mean_f1_scores = combined_results.groupby(['dataset', 'classifier'])['f1'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='f1', hue='classifier', data=mean_f1_scores)
plt.title('Mean F1 across classifiers')
plt.ylabel('Mean F1-score')
plt.xlabel('Classifiers')
plt.savefig('mean_f1_classifiers.png')
plt.show()
```

```python
mean_f1_scores_splitting = combined_results.groupby(['dataset', 'splitting'])['f1'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='f1', hue='splitting', data=mean_f1_scores_splitting)
plt.title('Mean F1-score by splitting strategy per dataset')
plt.ylabel('F1-score')
plt.xlabel('Datasets')
plt.savefig('f1_datasets_splitting.png')
plt.show()
```

```python
mean_f1_scores_sel = combined_results.groupby(['dataset', 'feature_selection'])['f1'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='f1', hue='feature_selection', data=mean_f1_scores_sel)
plt.title('Mean F1-score by feature selection method per dataset')
plt.ylabel('F1-score')
plt.xlabel('Datasets')
plt.savefig('f1_datasets_feature_selection.png')
plt.show()
```

```python
filtered_results = combined_results[combined_results['dataset'] != 'bank_marketing']
mean_f1_filtered_results = filtered_results.groupby(['dataset', 'scaling'])['f1'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='f1', hue='scaling', data=mean_f1_filtered_results)
plt.title('Mean F1-score by scaling method per dataset')
plt.ylabel('F1-score')
plt.xlabel('Datasets')
plt.savefig('f1_datasets_scaling.png')
plt.show()
```

```python
filtered_results = filtered_results[filtered_results['classifier'] == 'KNeighborsClassifier']
mean_f1_filtered_results = filtered_results.groupby(['dataset', 'scaling'])['f1'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='f1', hue='scaling', data=mean_f1_filtered_results)
plt.title('Mean F1-score by scaling method per dataset only for KNN')
plt.ylabel('F1-score')
plt.xlabel('Datasets')
plt.savefig('f1_datasets_scaling_onlyKNN.png')
plt.show()
```

```python
mean_time = combined_results.groupby(['dataset', 'classifier'])['training_time'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='training_time', hue='classifier', data=mean_time, log=True)
plt.title('Comparison of the training time across datasets and classifiers')
plt.ylabel('Training Time')
plt.xlabel('Datasets')
plt.savefig('trainingtime_datasets_classifiers.png')
plt.show()
```

```python
mean_time = combined_results.groupby(['dataset', 'classifier'])['preprocess_time'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='dataset', y='preprocess_time', hue='classifier', data=mean_time, log=True)
plt.title('Comparison of the preprocessing time across datasets and classifiers')
plt.ylabel('Preprocessing Time')
plt.xlabel('Datasets')
plt.savefig('preprocessingtime_datasets_classifiers.png')
plt.show()
```
